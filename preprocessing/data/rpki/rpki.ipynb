```python
# Load json file
import json
VRPS_JSON = "vrps.json"
data = json.load(open(VRPS_JSON))['roas']
print(f"Loaded {len(data)} VRPs from {VRPS_JSON}")
```

```python
# Count every asn in the vrps
from collections import defaultdict
asn_count = defaultdict(int)
for vrp in data:
    asn = vrp['asn']
    asn_count[asn] += 1

# Print the 10 most common asns
most_common_asns = sorted(asn_count.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top 10 ASNs by VRP count:")
for asn, count in most_common_asns:
    print(f"ASN {asn}: {count} VRPs") 

# Export to csv
import csv
with open("asn_vrp_counts.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["asn", "vri_count"])
    for asn, count in asn_count.items():
        writer.writerow([asn, count])
```
